## STANDARISASI 

```python
# duplicate dataset
df_standarirasi = df1.copy()
df_comparison = df1.copy()

# remove attribute type object
df_standarirasi.drop('name', axis=1, inplace=True)
df_standarirasi.drop('category', axis=1, inplace=True)
df_standarirasi.drop('link', axis=1, inplace=True)
df_standarirasi.drop('other_colors', axis=1, inplace=True)
df_standarirasi.drop('short_description', axis=1, inplace=True)
df_standarirasi.drop('designer', axis=1, inplace=True)
df_standarirasi.drop('sellable_online', axis=1, inplace=True)
df_standarirasi.drop('old_price', axis=1, inplace=True)
#standarisasi
standart_std = StandardScaler()
scaled_data = standart_std.fit_transform(df_standarirasi)

#menghapus atribut bertipe object dataset pembanding
df_comparison.drop('name', axis=1, inplace=True)
df_comparison.drop('category', axis=1, inplace=True)
df_comparison.drop('link', axis=1, inplace=True)
df_comparison.drop('other_colors', axis=1, inplace=True)
df_comparison.drop('short_description', axis=1, inplace=True)
df_comparison.drop('designer', axis=1, inplace=True)
df_comparison.drop('sellable_online', axis=1, inplace=True)
df_comparison.drop('old_price', axis=1, inplace=True)

print('Dataset sebelum scalling:',  # SEBELUM
      df_comparison.head(6),
      f'Nilai standar deviasi:\n{np.std(df_comparison)}',
      sep='\n', end='\n\n')

print('Dataset setelah scalling:',  # SESUDAH
      scaled_data[:6],
      f'Nilai standar deviasi: {np.std(scaled_data)}',
      sep='\n')
```
